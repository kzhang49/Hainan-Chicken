```python
import pandas as pd
import numpy as np

neighbourhood_type = ['Jamaica Plain','Back Bay','Dorchester','Fenway','South End','Allston','Downtown','Brighton',
                 'Beacon Hill','South Boston','East Boston','Roxbury','North End','Charlestown','Mission Hill',
                 'Chinatown','Roslindale','West Roxbury','South Boston Waterfront','West End','Hyde Park','Mattapan',
                 'Bay Village','Leather District','Longwood Medical Area']
room_type = ['Entire home/apt','Private room','Shared room']
property_type = ['Apartment','House','Condominium','Other','Townhouse','Loft','Bed & Breakfast','Boat','Guest suite',
                 'Villa','Boutique hotel','In-law','Guesthouse','Hostel','Timeshare','Dorm','Serviced apartment']
tf = ['t','f']
cancellation_policy_type = ['strict', 'moderate', 'flexible', 'super_strict_30', 'super_strict_60']
host_response_time_type = ['within an hour','within a few hours','within a day','a few days or more','0']

def main():
    print("\nAirbnb Price prediction model by Hainan Chicken!")
    print("Follow the instruction below, you would get a price suggestion for your lovely house...")

    neighbourhood_type = neighbourhood()
    room_type = room()
    property_type = prop()
    accommodates_num = accommodates()
    ib_tf = instant_bookable()
    bedrooms_num = bedrooms()
    bathrooms_num = bathrooms()
    beds_num = beds()
    sum_of_nights = nights()
    pets_allowed_tf = pets_allowed()
    gym_tf = gym()
    Washer_dryer_tf = Washer_dryer()
    sum_of_amenities = amenities()
    cancellation_policy_type = cancellation_policy()
    host_response_time_type = host_response_time()

    d = {'host_response_time': [host_response_time_type], 'host_response_rate': [100], 'neighbourhood_cleansed': [neighbourhood_type],
         'property_type': [property_type], 'room_type': [room_type], 'accommodates': [accommodates_num], 'bathrooms': [bathrooms_num],
         'bedrooms': [bedrooms_num],'beds':[beds_num], 'pets_allowed':[pets_allowed_tf],'Washer_dryer':[Washer_dryer_tf],
         'gym':[gym_tf], 'sum_of_amenities':[sum_of_amenities],	'instant_bookable':[ib_tf],'cancellation_policy':[cancellation_policy_type],
         'sum_of_nights':[sum_of_nights], 'price':[0]}
    df = pd.DataFrame(data=d)
    
    return df
    #print("\n We suggest you set your price at:... ")


def host_response_time():
    while True:
        hr_in = input("\n- What is your response time? (eg.within an hour, within a few hours, within a day, a few days or more, 0): ")
        if hr_in in host_response_time_type:
            break
        else:
            print("\n > Sorry, please enter a response time.")
    return hr_in


def cancellation_policy():
    while True:
        cp_in = input("\n- What is your cancellation policy? (eg.strict, moderate, flexible, super_strict_30, super_strict_60): ")
        if cp_in in cancellation_policy_type:
            break
        else:
            print("\n > Sorry, please enter a cancellation policy.")
    return cp_in


def nights():
    while True:
        nights_in = input("\n- How many nights available in one year? ")
        nights = int(nights_in)
        break
    else:
        print("\n > Sorry, please enter a valid number.")
    return nights


def room():
    while True:
        room_in = input("\n- What type of room do you have? (eg.Entire home/apt, Private room, Shared room): ")
        if room_in in room_type:
            break
        else:
            print("\n > Sorry, please enter a valid room type.")
    return room_in


def prop():
    while True:
        property_in = input("\n- What type of property do you have? (eg.Apartment, House, Other): ")
        if property_in in property_type:
            break
        else:
            print("\n > Sorry, please enter a valid room type.")
    return property_in


def neighbourhood():
    while True:
        neighbourhood_in = input("\n- Which neighbourhood does your property locate? (eg.Jamaica Plain, Back Bay, Dorchester): ")
        if neighbourhood_in in neighbourhood_type:
            break
        else:
            print("\n > Sorry, please enter a valid neighbourhood.")
    return neighbourhood_in


def pets_allowed():
    while True:
        pa_in = input("\n- Is pet allowed? (0/1): ")
        if pa_in in ['0','1']:
            break
        else:
            print("\n > Sorry, please enter a valid number.")
    return int(pa_in)


def gym():
    while True:
        g_in = input("\n- Is there a gym? (0/1): ")
        if g_in in ['0','1']:
            break
        else:
            print("\n > Sorry, please enter a valid number.")
    return int(g_in)


def Washer_dryer():
    while True:
        wd_in = input("\n- Is there a Washer dryer? (0/1): ")
        if wd_in in ['0','1']:
            break
        else:
            print("\n > Sorry, please enter a valid number.")
    return int(wd_in)


def instant_bookable():
    while True:
        ib_in = input("\n- Is your property instant bookable? (t/f): ")
        if ib_in in tf:
            break
        else:
            print("\n > Sorry, please enter a valid character.")
    return ib_in


def accommodates():
    while True:
        accommodates_in = input("\n- How many people could your property accommodate? ")
        accommodates = int(accommodates_in)
        break
    else:
        print("\n > Sorry, please enter a valid number.")
    return accommodates


def bedrooms():
    while True:
        bedrooms_in = input("\n- How many bedrooms? ")
        bedrooms = int(bedrooms_in)
        break
    else:
        print("\n > Sorry, please enter a valid number.")
    return bedrooms


def bathrooms():
    while True:
        bathrooms_in = input("\n- How many bathrooms? ")
        bathrooms = int(bathrooms_in)
        break
    else:
        print("\n > Sorry, please enter a valid number.")
    return bathrooms


def beds():
    while True:
        beds_in = input("\n- How many beds? ")
        beds = int(beds_in)
        break
    else:
        print("\n > Sorry, please enter a valid number.")
    return beds


def amenities():
    while True:
        amenities_in = input("\n- How many amenities? ")
        amenities = int(amenities_in)
        break
    else:
        print("\n > Sorry, please enter a valid number.")
    return amenities
```

```python
df_test_data = main()
df_test_data
```

## Split dataframe to features(X) & class(y)

## Feature Engineering

### Elminate Unimportant Features

```python
df_final = df.drop(['wifi', 'heating', 'is_location_exact'], axis = 1)
X = df_final.values
df_final.shape
```

### One-Hot-Encoding

```python
df_1hot = df_final.copy()
X_dummies = pd.get_dummies(df_1hot, columns=df_1hot.columns, drop_first=True) #one hot encoding
X_dummies.dtypes
```

## Split dataset to X_train, X_test, y_train, y_test

```python
df_user=[X_dummies.iloc[-1].values]
X_dummies.drop(X_dummies.tail(1).index, inplace=True)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dummies, new_y, random_state=0)
```

## Start Machine Learning...

```python
def report_results(grid):
    cv_keys = ('mean_test_score', 'std_test_score', 'params')
    print('mean_test_score +/ std_test_score, {params}') 
    for r, _ in enumerate(grid.cv_results_['mean_test_score']):
        bf = '*' if grid.cv_results_[cv_keys[0]][r]==grid.best_score_ else ' '
        print(bf+"%0.3f +/- %0.2f %r"
          % (grid.cv_results_[cv_keys[0]][r],
             grid.cv_results_[cv_keys[1]][r] / 2.0,
             grid.cv_results_[cv_keys[2]][r]))
```

### Random Forest

```python
from sklearn.cross_validation import cross_val_score
from sklearn.model_selection import cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth=12, n_estimators = 2000, random_state=5)
rf.fit(X_train,y_train)
scores = cross_val_score(rf, X_train, y_train, cv=10, scoring='accuracy')
print(scores)
print("mean score: ", scores.mean())
```

### Ada Boost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ab = AdaBoostClassifier(n_estimators=25, random_state=7)
ab.fit(X_train,y_train)
scores = cross_val_score(ab, X_train, y_train, cv=10, scoring='accuracy')
print(scores)
print("mean score: ", scores.mean())
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier(max_depth=6, max_features=0.3, min_samples_split=4)
gb.fit(X_train,y_train)
scores = cross_val_score(gb, X_train, y_train, cv=10, scoring='accuracy')
print(scores)
print("mean score: ", scores.mean())
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(class_weight = 'balanced', C = 0.9, multi_class = 'multinomial',
                           solver = 'newton-cg')
logreg.fit(X_train,y_train)
scores = cross_val_score(logreg, X_train, y_train, cv=10, scoring='accuracy')
print(scores)
print("mean score: ", scores.mean())
```

### Ensemble Method

```python
from sklearn.ensemble import VotingClassifier
clf = VotingClassifier(estimators=[('RF', rf), ('AB', ab), 
                                   ('GB', gb), ('LG', logreg)],
                        voting='soft', weights = [1.5, 1, 2, 0.5])
clf.fit(X_train,y_train)
ac = cross_val_score(clf, X_train, y_train, cv=5, scoring = 'accuracy')
print('mean_test_score = ', ac.mean())
print('std_test_score = ', ac.std())
```
